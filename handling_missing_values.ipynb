## How to handle missing values (Nulls) in you dataset

```python
import pandas as pd 
```

```python
# create the data frame 
df = pd.read_csv("Bengaluru_House_Data.csv")
df.head(10)
```

```python
df.info()
```

```python
# the shape of the data frame 
df.shape
```

```python
# check the null values 

df.isnull()
```

```python
# check the number of null values for every column in your dataset : 
df.isnull().sum()
```

```python
# check the number of null values in your dataset 
print(f"Number of null values in the dataframe: {df.isnull().sum().sum()}")
```

### filling Null values with fillna() method

```python
# let's say we need to fill all null values with initial value 0:
df1 = df.fillna(value=0)
df1.head(10)
```

```python
# check the number of nulls in each col:
df1.isnull().sum()
```

```python
# Original data frame with null values 
df.head(10)
```

```python
# Filling Null values with previous ones : 
df2 = df.fillna(method="pad")
df2.head(10)
```

```python
# original df :
df.head(10)
```

```python
# filling nulls with next ones : 
df3 = df.fillna(method="bfill")
df3.head(10)
```

```python
# original df
df.head(10)
```

```python
# filling Nulls with different values in different columns 
df4 = df.fillna({"society":"abcd",
                 "balcony" : 5})  

df4.head(10)
```

```python
# original df: 
df.head(10)
```

```python
# To fill nulls with the mean of the numeric columns :
df5 = df 
df5["balcony"].fillna(df5["balcony"].mean() , inplace= True ) # --> we can use max and min too 
df5["society"].fillna(method="pad" , inplace= True)
df5.head(10)
```

### handling with dropna() method 

```python
# making a new df : 
import numpy as np

dict = {'First Score':[100, 90, 20, 95],
        'Second Score': [30, 45, 56, 30],
        'Third Score':[55, 40, np.nan, 98],
        'fourth Score':[20,np.nan , 44 , 70],
        'fifth Score':[60,30 ,10, 70]}

df_new = pd.DataFrame(dict)
df_new
```

```python
# drop rows with at least one Nan value (Null value)
df2 = df_new.dropna()
df2
```

```python
#  Dropping rows if all values in that row are missing
df3 = df_new.dropna(how='all')
df3
```

### Filling null values with replace() method 


```python
df_new
```

```python
df4 = df_new.replace(to_replace=np.nan , value= -5)
df4
```

### Filling Nulls with interpolate() method

```python
df_new
```

```python
# interpolate the missing values using Linear method
df5 = df_new.interpolate(method="linear" , limit_direction="forward")
df5
```
